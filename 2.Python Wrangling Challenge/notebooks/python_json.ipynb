## Python and JSON

#### 1. Imports ```json``` which is part of the core Python library. Used to handle [JSON](https://www.w3schools.com/python/python_json.asp) data.
> Here we'll be using plain Python to handle some JSON data.

````
import json
````

#### 2. Open the `swimming_psb_data.json` file and deserialize the JSON data with ```json.load()```.
> You should end up with a variable assigned to a list of dictionaries.

````
## Open file using ' 's and correct file path
open_raw_json = open('/work/2.Python Wrangling Challenge/data/swimming_psb_data.json')

## Deserialize 
list_of_swimmers = json.load(open_raw_json)

````

#### 3. Filter list to contain only dictionaries belonging to Great British Athletes only.

````
## create new list, cycle through the json to find Great Britain and append to new list
British_Athletes = []
for British in list_of_swimmers: 
    if British ['c_NOC'] == 'Great Britain':
        British_Athletes.append(British)

print (British_Athletes)
````

#### 4. Serialize data with ```json.dumps()``` and write the data to a new file called `gb_swimming_psb_data.json` within the `data` folder.

````
with open('/work/2.Python Wrangling Challenge/data/gb_swimming_psb_data.json', 'w') as newfile:
    British_Athletes_json = json.dumps(British_Athletes) 
    newfile.write(British_Athletes_json)
````
